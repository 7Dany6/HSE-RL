```python
import pandas as pd
import numpy as np
from scipy.stats import kstest, norm, beta, entropy
import matplotlib.pyplot as plt
import random
from scipy.special import kl_div
from collections import Counter
from scipy.stats import beta
import operator
import seaborn as sns
```

# CODE FOR THE CONFERENCE

```python
a, b = 2.31, 0.627
x = np.linspace(beta.ppf(0.01, a, b),
                beta.ppf(0.99, a, b), 100)
```

```python
batches = 10
batch_size = 100
batch_number = 0
```

```python
best_medicine = 0
medicine = 0
```

```python
kstests = [1]
data = {}
data['0'] = [0] * 10 + [1] * 90
data['1'] = [0] * 20 + [1] * 80
num_of_responses = [0] * 2
num_of_rejections = [0] * 2
medicine_selected = []
total_responses = 0
batch_number = 0
for _ in range(batches):
    batch_number += 1
    if kstests[-1] < 0.05:
        best_medicine = max(intermediate_results.items(), 
                            key=operator.itemgetter(1))[0]
        print(f"The best medicine is {best_medicine}, stop after batch #{batch_number-1}")
        break
    for _ in range(batch_size):
        intermediate_results = {}
        treatment_A = beta(num_of_responses[0] + 1,
                           num_of_rejections[0] + 1)
        treatment_B = beta(num_of_responses[1] + 1,
                           num_of_rejections[1] + 1)
        intermediate_results['A'], intermediate_results['B'] = treatment_A.mean(), treatment_B.mean()
        A_entropy, B_entropy = entropy(treatment_A.pdf(x)), entropy(treatment_B.pdf(x))
        print(f"Batch number {batch_number}: {treatment_A.mean()}, {treatment_B.mean()}")
        if A_entropy > B_entropy:
            medicine = 0
        elif B_entropy > A_entropy:
            medicine = 1
        medicine_selected.append(medicine)
        #response = random.randint(0,1)
        response = random.choice(data[str(medicine)])
        if response == 1:
            num_of_responses[medicine] += 1
        else:
            num_of_rejections[medicine] += 1
        total_responses += response
    kstests.append(kstest(treatment_A.cdf(x), treatment_B.cdf(x))[1])
```
